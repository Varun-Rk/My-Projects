```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Price Index Data from fred economic data
```

```python
Prices = pd.read_csv('Downloads/CSUSHPISA-2.csv')
```

```python
Prices
```

```python
Prices.info()
```

```python
#Converting dates from object to datetime
```

```python
Prices['DATE'] = pd.to_datetime(Prices['DATE'])
```

```python
Prices =  Prices.resample('Q').mean()
```

```python
# New homes for sale in Us 
```

```python
home_fs = pd.read_csv('Downloads/HNFSEPUSSA-2.csv')
```

```python
home_fs.rename(columns={'HNFSEPUSSA':'homes_sold'},inplace=True)

home_fs['DATE'] = pd.to_datetime(home_fs['DATE'])

home_fs.set_index('DATE',inplace=True)

home_fs = home_fs.resample('Q').mean()

home_fs.drop(home_fs.index.min(),inplace=True)
```

```python
home_fs
```

```python
new_houses = pd.read_csv('Downloads/MSACSR-2.csv')
new_houses
```

```python
new_houses.rename(columns={'MSACSR':'new_homes'},inplace=True)

new_houses['DATE'] = pd.to_datetime(new_houses['DATE'])

new_houses.set_index('DATE',inplace=True)

new_houses = new_houses.resample('Q').mean()

new_houses.drop(new_houses.index.min(),inplace=True)
```

```python
new_houses
```
